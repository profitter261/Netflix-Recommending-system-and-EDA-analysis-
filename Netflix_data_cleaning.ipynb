Exploration Summary


*   we have a dataset of size of 1000 x 9 size
*   Dataset seems to be a clean dataset with no duplicate values




```python
df['release_date'] = pd.to_datetime(df['release_date'])
df.head()
```

```python
df['release_date'].dtypes
```

```python
df['release_date'] = df['release_date'].dt.year
df.head()
```

```python
df['release_date'].dtypes
```

```python
cols = ['overview','original_language']
df.drop(cols,axis=1,inplace=True)
```

```python
df.columns
```

```python
df.head()
```

```python
for i in df.index:
  if df.loc[i,'vote_average'] >= 8.0:
    df.loc[i,'vote_average'] = 'Popular'
  elif df.loc[i,'vote_average'] >= 7.0 and df.loc[i,'vote_average'] < 8.0:
    df.loc[i,'vote_average'] = 'above average'
  elif df.loc[i,'vote_average'] >= 6.0 and df.loc[i,'vote_average'] < 7.0:
    df.loc[i,'vote_average'] = 'below average'
  else:
    df.loc[i,'vote_average'] = 'non-popular'
```

```python
df.head()
```

```python
df['vote_average'].value_counts()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df['genre'] = df['genre'].str.split(',')
df = df.explode('genre').reset_index(drop=True)
```
